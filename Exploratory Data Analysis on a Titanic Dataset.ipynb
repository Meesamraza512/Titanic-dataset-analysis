# Handle the missing values

```python
# Create a copy of the dataframe
df_clean = df.copy()

# Handling missing values:
# Fill missing Age values with the median age
age_median = df_clean['Age'].median()
df_clean['Age'].fillna(age_median, inplace=True)

# Fill missing Embarked values with the mode
embarked_mode = df_clean['Embarked'].mode()[0]
df_clean['Embarked'].fillna(embarked_mode, inplace=True)

# For Cabin, since a large portion is missing, we fill with 'Unknown'
df_clean['Cabin'].fillna('Unknown', inplace=True)
```

# Check the Missing values

```python
# Check the Missing values
print("\nMissing values:")
print()

df_clean.isnull().sum()
```

# Distribution of Numerical data type columns

```python
#Distribution of Numerical data type columns

df_clean.hist(figsize=(10, 8), bins=20)
plt.show()
```

# Distribution of Age

```python
# Plot: Distribution of Age
plt.figure(figsize=(9, 7))
plt.hist(df_clean['Age'], bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

#### The age distribution shows most passengers were between 20-40 years old, with a right-skewed distribution.

# Distribution of Fare

```python
# Plot: Distribution of Fare
plt.figure(figsize=(9, 7))
plt.hist(df_clean['Fare'].dropna(), bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

```

#### The fare distribution is heavily right-skewed, with most passengers paying lower fares and a few paying significantly higher amounts.

# Distribution of Sex

```python
# Pie chart for 'Sex' column distribution
sex_counts = df['Sex'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.0f%%', startangle=140)
plt.title('Distribution of Sex Classes')
plt.show()
```

#### The chart shows that male passengers significantly outnumbered female passengers on the Titanic, with males making up approximately 65% of all passengers and females about 35%.

# Survival Rate (Target Variable) Count Plot:

```python
# Survival Rate (Target Variable) Count Plot:
plt.figure(figsize=(9, 7))

p=sns.countplot(x='Survived', data=df_clean)
plt.title("Survival Count")


# Show data labels
for bars in p.containers:
    p.bar_label(bars)

plt.show()
```

#### Out of 891 passengers, more people did not survive (549) than those who survived (342).

```python
# Bar plot for Passenger Class distribution
plt.figure(figsize=(9, 7))

a=sns.countplot(x='Pclass', data=df_clean)
plt.title('Passenger Class Distribution')
plt.xlabel('Pclass')
plt.ylabel('Count')

# Show data labels
for bars in a.containers:
    a.bar_label(bars)

plt.show()
```

#### Most passengers were traveling in 3rd class, followed by 1st class, and then 2nd class.

# Correlation matrix 

```python
# Correlation matrix 
plt.figure(figsize=(10, 8))
corr_matrix = df_clean.corr(numeric_only=True)

# plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

#### Positive correlation between Survival and Fare (0.26)
#### Negative correlation between Survival and Pclass (-0.34)
#### Negative correlation between Pclass and Fare (-0.55)

# Visualizing the Distribution of Age and Fare:

```python
# Visualizing the Distribution of Age:
plt.figure(figsize=(10, 8))

sns.boxplot(x='Survived', y='Age', data=df_clean)
plt.title("Age vs Survival")
plt.show()


# Visualizing the Distribution of Fare:
plt.figure(figsize=(10, 8))

sns.boxplot(x='Survived', y='Fare', data=df_clean)
plt.title("Fare vs Survival")
plt.show()
```

#### While age might show some differences between groups, ticket fare tends to exhibit sharper disparities between survivors and non-survivors. This implies that economic status (as inferred from the fare) might have had a more pronounced impact on who survived the disaster.
